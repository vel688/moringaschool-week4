```python
#removing outliers using interquartile range
Q1=cars.quantile(0.25)
Q3=cars.quantile(0.75)

IQR=Q3-Q1


#find out where exactly the outliers are in the dataframe
((cars < Q1 - 1.5 * IQR) | (cars > Q3 + 1.5 * IQR))
```

```python
#remove the outliers
cars = cars[ ~(( cars < Q1 - 1.5 * IQR) | (cars > Q3 + 1.5 * IQR)) .any (axis=1)]
```
